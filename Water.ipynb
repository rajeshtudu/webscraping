```python
url = 'https://www.worldometers.info/water/'
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

html_content = requests.get(url).content
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.find('tbody')
tr = soup.find_all('tr')

all_data = []
for i in range(1, len(tr)):
    water = []
    td = tr[i].find_all('td')
    for data in td:
        water.append(data.text)
    all_data.append(water)
    
clean_data = list(map(lambda x: x[0:4], all_data))
column_name = ['Country', 'Yearly Water Used', 'Daily Water Used', 'Population']

df = pd.DataFrame(clean_data, columns = column_name)
```
